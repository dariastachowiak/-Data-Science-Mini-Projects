```python
#Group months by year & find average price per year
```

```python
df['Year']=df['Month'].apply(lambda t: t.year)
```

```python
df.tail()
```

```python
df.shape
```

```python
dfg=df.groupby(by=['London_Borough','Year']).mean()
```

```python
dfg=dfg.reset_index()
```

```python
dfg.head()
```

```python
#Define function that calculates price change ratio between 1998 and 2018
```

```python
def create_price_ratio(d):
    y1998=float(d['Average_price'][d['Year']==1998])
    y2018=float(d['Average_price'][d['Year']==2018])
    ratio=[y2018/y1998]
    return ratio
```

```python
#tested out function on specific borough
```

```python
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])

                   
```

```python
#Created dictionary for finalized data that will include the borough and the price change ratio
```

```python
final = {}
```

```python
for b in dfg['London_Borough'].unique():
    borough=dfg[dfg['London_Borough']==b]
    final[b]=create_price_ratio(borough)
```

```python
print(final)
```

```python
#Turned dictionary into DataFrame
```

```python
df_ratios=pd.DataFrame(final)
```
